# Possible values of n and o for MWA

## Finding the n and o values

```python
T = 1829  # Number of SNe
N = 3   # Number of bins

# User inputs for the range of n
min_n = 50
max_n = 700

# List to store valid (n, o) pairs
valid_combinations = []

# Iterate through the range of n values
for n in range(min_n, max_n + 1):  
    o = ((n * N) - T) / (N - 1)
    
    if o.is_integer() and o >= 0:  # Ensure o is a non-negative integer
        valid_combinations.append((n, int(o)))

# Print results
if valid_combinations:
    print("\nPossible (n, o) values:")
    for n, o in valid_combinations:
        print(f"n = {n}, o = {o}")
else:
    print("\nNo valid (n, o) values found in the given range.")
```

## Bin seperation

```python
def define_bins(start, stop, o, bins, n):
    bin_ranges = []
    
    for i in range(bins):
        bin_start = start + i * (n - o)
        bin_end = bin_start + n - 1
        
        if bin_end > stop:
            bin_end = stop  # Ensure the last bin does not exceed the stop value
        
        bin_ranges.append((bin_start, bin_end))
    
    return bin_ranges

# Example usage
start = 1
stop = 1829
n = 649
o = 59
bins = 3


ranges = define_bins(start, stop, o, bins, n)
print("Bin Ranges:")
for i, (low, high) in enumerate(ranges, 1):
    print(f"Bin {i}: ({low}-{high})")
```
