# 1- sigma Compatibility checker

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline
```

```python
import os
print("Current Working Directory:", os.getcwd())
print("Files in Directory:", os.listdir())
# u can uncomment this if you are sure about where you are 
```

```python
 # Read the .txt file
file_path = "waCDM_12bins_JLA_varyingOmegam.txt" 
data = pd.read_csv(file_path, delim_whitespace=True)  # Assuming whitespace separation

file_name = os.path.splitext(os.path.basename(file_path))[0]  # Extracts 'wCDM_3_bins' from 'wCDM_3_bins.txt'

# Print the entire file as a table
print("\nFull Data Table:\n")
print(data.to_string(index=False))  # Prints the table without the BIN
```

```python
# Define the parameters of interest
params = ["Om", "Omerr", "w0", "w0err", "wa", "waerr", "w", "werr"]  

# Find which parameters are present in the file
present_params = [param for param in params if param in data.columns]  

# Print the detected parameter names
print("\nDetected Parameters:")
print(present_params)

# Extract and print the values as a table
if present_params:
    param_table = data[present_params]  # Extract only present parameters
    print("\nParameter Values Table:")
    print(param_table.to_string(index=False))  # Print table without row indices
else:
    print("\nNo relevant parameters found in the file.")
```

### Omega_m (Om) weighted average

```python
# Extract Om values and their 1-sigma errors
Om_values = data["Om"]  
sigma_Om = data["Omerr"] 

#Compute 1-Sigma Range
Om_min = Om_values - sigma_Om
Om_max = Om_values + sigma_Om

# Check 1-Sigma Compatibility
compatibility = (Om_values >= Om_min) & (Om_values <= Om_max)
print("\nOmega_m values:\n", Om_values.to_list())
print("\n1-Sigma Lower Bound:\n", Om_min.to_list())
print("\n1-Sigma Upper Bound:\n", Om_max.to_list())
print("\n1-Sigma Compatibility Check:\n", compatibility.to_list())
```

```python
# Extract compatible values and compute the weighted average
compatible_Om = Om_values[compatibility]
compatible_sigma = sigma_Om[compatibility]

if len(compatible_Om) > 0:
    weights = 1 / (compatible_sigma ** 2)  # Weight = 1 / (error^2)
    weighted_avg_Om = np.sum(weights * compatible_Om) / np.sum(weights)
    weighted_error_Om = np.sqrt(1 / np.sum(weights))  # Standard weighted error
    print(f"\nWeighted Average Omega_m: {weighted_avg_Om:.5f} ± {weighted_error_Om:.5f}")
else:
    print("\nNo compatible Omega_m values found within 1-sigma.")
```

```python
# Plot Omega_m with 1-sigma bounds
indices = np.arange(1, len(Om_values) + 1)  # Start from 1 instead of 0
plt.figure(figsize=(8, 5))
plt.errorbar(indices, Om_values, yerr=sigma_Om, fmt='o', color='black',
             ecolor='blue', elinewidth=1, capsize=4, label=r"$\Omega_m$ Values with 1$\sigma$ Error Bars")
#plt.fill_between(range(len(Om_values)), Om_min, Om_max, color='gray', alpha=0.3, label="1-Sigma Region")

# Mark the weighted average if computed
if len(compatible_Om) > 0:
    plt.axhline(weighted_avg_Om, color='red', linestyle='--', label=f"Weighted Avg: {weighted_avg_Om:.5f} ± {weighted_error_Om:.5f}")

# Set all indices explicitly on x-axis (starting from 1)
plt.xticks(indices, labels=[str(i) for i in indices])

plt.xlabel("BIN")
plt.ylabel(r"Matter Density Parameter ($\Omega_m$)")
plt.title(r"1-Sigma Compatibility of $\Omega_m$")
plt.legend()
#plt.grid(True)

# Save the plot as a PDF using the same file name
pdf_filename = f"Om_vs_binnumber {file_name}.pdf"  
plt.savefig(pdf_filename, format="pdf", bbox_inches="tight")
print(f"\nPlot saved as: {pdf_filename}")

# Save plot as png
#plt.savefig('parameter_vs_binnumber.png', bbox_inches='tight')
png_filename = f"Om_vs_binnumber {file_name}.png" 
plt.savefig(png_filename, format="png", bbox_inches="tight")
print(f"\nPlot saved as: {png_filename}")
plt.show()
```
